```python
import os
import shutil
from pathlib import Path

# Path to your kaggle.json file (update this if needed)
notebook_path: Path = Path().resolve()
root_dir: Path = notebook_path.parent 
kaggle_json_path: Path = root_dir/"kaggle.json"

if not kaggle_json_path.exists():
    print("Please place your kaggle.json file in the current directory.")
else:
    # Create the .kaggle directory in the user's home if it doesn't exist
    kaggle_dir = Path.home() / ".kaggle"
    kaggle_dir.mkdir(parents=True, exist_ok=True)
    
    # Copy kaggle.json to the .kaggle directory
    shutil.copy(kaggle_json_path, kaggle_dir / "kaggle.json")
    
    # Set permissions to read/write for user only
    os.chmod(kaggle_dir / "kaggle.json", 0o600)
    
    print(f"kaggle.json moved to {kaggle_dir} with correct permissions.")
```

Download dataset from Kaggle

```python
import zipfile
from pathlib import Path
from kaggle.api.kaggle_api_extended import KaggleApi

# Initialize Kaggle API
api = KaggleApi()
api.authenticate()

# Dataset info
dataset_name = "fahadullaha/facial-emotion-recognition-dataset"
zip_filename = "facial-emotion-recognition-dataset.zip"


data_dir = root_dir / "data"
fer_dir = data_dir / "fer_dataset"
fer_dir.mkdir(parents=True, exist_ok=True)

# Download dataset (without unzipping)
print("Downloading dataset...")
api.dataset_download_files(dataset_name, path=data_dir, unzip=False)

# Unzip dataset
zip_path = data_dir / zip_filename
print(f"Unzipping {zip_path}...")
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(fer_dir)

# Optionally, remove the zip file after extraction
zip_path.unlink()

# List files
fer_processed_dir: Path = fer_dir/"processed_data"
print("Files in dataset directory:")
for f in fer_processed_dir.iterdir():
    print(f.name)
```
